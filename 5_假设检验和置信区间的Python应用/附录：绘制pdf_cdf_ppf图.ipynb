# 计算和绘制正态分布的p.d.f.、c.d.f和p.p.f图

## Scipy
是python的一个科学计算库，可以进行线性代数、数值计算、信号处理，优化和统计分析。下图是scipy的功能列表：

<img src="image/scipy.png" alt="scipy" style="zoom:50%;" />

计量经济学使用的scipy.stats工具，可以计算各类分布的概率密度pdf，计算各类统计量等等。

## 例子：使用scipy绘制标准正态分布概率密度图（p.d.f.)

```python
%matplotlib inline
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
xs = np.linspace(-3, 3, 200)
normal = stats.norm.pdf(xs)
plt.figure(figsize=(10,6))
plt.plot(xs, normal, label='p.d.f. of standard normal dist')
plt.legend()
plt.savefig('图片/norm_pdf.jpg')
```

## 例子：使用scipy绘制标准正态分布累积分布图（c.d.f.）
cdf是概率的累积分布，即小于等于某个值出现的概率总和。

```python
%matplotlib inline
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
xs = np.linspace(-3, 3, 200)
normal = stats.norm.cdf(xs)
plt.figure(figsize=(10,6))
plt.plot(xs, normal, label='c.d.f of standard normal dist')
plt.legend()
plt.savefig('图片/norm_cdf.jpg')
```

## 例子：使用scipy绘制标准正态分布Percent Point Function（p.p.f.）
ppf是cdf的逆运算，也就是已知概率总和，求对应的统计量的值

```python
%matplotlib inline
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
xs = np.linspace(0, 1, 200)
normal = stats.norm.ppf(xs)
plt.figure(figsize=(10,6))
plt.plot(xs, normal, label='p.p.f of standard normal dist')
plt.legend()
plt.savefig('图片/norm_ppf.jpg')
```
